```python
import gym
import numpy as np
from stable_baselines3 import A2C

# Create the gym environment
env = gym.make('Poppy-v0')

# Set the random seed for reproducibility
seed = 123
env.seed(seed)
np.random.seed(seed)

# Define the A2C agent with custom hyperparameters
learning_rate = 0.1  # Change the learning rate
policy_architecture = 'MlpPolicy'  # Change the policy architecture
entropy_coef = 0.01  # Change the entropy coefficient

model = A2C(
    policy=policy_architecture,
    env=env,
    learning_rate=learning_rate,
    policy_kwargs=dict(activation_fn=torch.nn.ReLU),  # Example: change activation function
    verbose=1,
    ent_coef=entropy_coef,
)

# Train the agent
total_timesteps = 10000
model.learn(total_timesteps=total_timesteps)

# Save the trained model
model.save("a2c_poppy_env")

# Load the trained model
model = A2C.load("a2c_poppy_env")

# Evaluate the trained model
eval_episodes = 10
for _ in range(eval_episodes):
    obs = env.reset()
    done = False
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, _ = env.step(action)
        env.render()

# Close the gym environment
env.close()
```
